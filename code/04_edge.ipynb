```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io, util, color, transform, filters, feature
```

```python
img = util.img_as_float32( color.rgb2gray( io.imread('data/Mines_Teddy_Bear.png') ) )

plt.set_cmap('gray') # set default colormap to 'gray'
plt.imshow(img)
```

### Sobel filter & edges

```python
edge_sobel_h = filters.sobel_h(img)
edge_sobel_v = filters.sobel_v(img)

fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(8,4))
ax[0].imshow(edge_sobel_h)
ax[1].imshow(edge_sobel_v)
```

```python
edge_amp = np.sqrt( (edge_sobel_h**2 + edge_sobel_v**2) )
plt.imshow(edge_amp)
```

### Canny edge detector

```python
edge_canny_1 = feature.canny(img, sigma=1)
edge_canny_2 = feature.canny(img, sigma=2)
edge_canny_3 = feature.canny(img, sigma=3)

fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(12,4))
ax[0].imshow(edge_canny_1)
ax[1].imshow(edge_canny_2)
ax[2].imshow(edge_canny_3)
```
